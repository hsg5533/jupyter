```python
url = "https://www.airkorea.or.kr/web/pmRelay?itemCode=10007&pMENU_NO=108"

driver = webdriver.Chrome('chromedriver')

driver.implicitly_wait(5)

driver.get(url)
time.sleep(1)
```

```python
link_day = driver.find_element(By.CSS_SELECTOR, '#strDateDiv2')
link_year = Select(driver.find_element(By.CSS_SELECTOR, "#searchDate_yyyy"))

mon = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
city = ['02', '031', '032', '033', '041', '042', '043', '051', '052', '053', '054', '055', '061', '062', '063', '064', '044']

search_btn = driver.find_element(By.CSS_SELECTOR, "#cont_body > form > div > div > a:nth-child(1)")
down_btn = driver.find_element(By.CSS_SELECTOR, "#cont_body > form > div > div > a.xls")

link_day.click()
link_year.select_by_value('2021')

link_mon = Select(driver.find_element(By.CSS_SELECTOR, "#searchDate_mm"))
link_mon.select_by_value('06') #이거 바꿔가면서 월별로 다운로드 01, 02, 03, 04, 05, 06, 07, 08, 09, 10, 11, 12
for c in city:
    link_c = Select(driver.find_element(By.CSS_SELECTOR, "#district"))
    link_c.select_by_value(c)
    print(link_c)
    time.sleep(1)
    search_btn.click()
    driver.implicitly_wait(10)
    down_btn.click()
    driver.implicitly_wait(10)
```

# **2022년 일별(24시간) 데이터 크롤링**
- 부산 6월 시간별 미세먼지 수치 크롤링

```python
url = "https://www.airkorea.or.kr/web/pmRelay?itemCode=10007&pMENU_NO=108"

driver = webdriver.Chrome('chromedriver')

driver.implicitly_wait(5)

driver.get(url)
time.sleep(1)
```

```python
city = ['02', '031', '032', '033', '041', '042', '043', '051', '052', '053', '054', '055', '061', '062', '063', '064', '044']

search_btn = driver.find_element(By.CSS_SELECTOR, "#cont_body > form > div > div > a:nth-child(1)")
down_btn = driver.find_element(By.CSS_SELECTOR, "#cont_body > form > div > div > a.xls")

pre_btn = driver.find_element(By.CSS_SELECTOR, "#dateDiv_1 > span > a.arrow_right")

for i in range(1, 31, 1):
    for c in city:
        link_c = Select(driver.find_element(By.CSS_SELECTOR, "#district"))
        link_c.select_by_value(c)
        print(link_c)
        time.sleep(1)
        search_btn.click()
        driver.implicitly_wait(10)
        down_btn.click()
        driver.implicitly_wait(10)
    pre_btn.click()
    driver.implicitly_wait(5)
```
