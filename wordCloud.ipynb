```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np
```

```python
with open('data/대한민국헌법.txt', 'r', encoding='utf-8') as f:
    text = f.read()

okt = Okt() # 형태소 분석기
nouns = okt.nouns(text) # 명사만 추출
words = [n for n in nouns if len(n) > 0] # 단어의 길이가 1개인 것은 제외
c = Counter(words) # 위에서 얻은 words를 처리하여 단어별 빈도수 형태의 딕셔너리 데이터를 구함
```

```python
wc = WordCloud(font_path='malgun', width=1080, height=1080, background_color='white',scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(c)
plt.imshow(gen)
plt.axis('off')
plt.show()
```

```python
wc.to_file('data/법전_워드클라우드.png')
```

```python
f=open("data/대한민국헌법.txt",encoding='utf-8')
data=f.read()
npl=Okt()   # 형태소분석기Okt생성
nouns=npl.nouns(data)# 형태소 분석기로 단어추출
count=Counter(nouns) # 단어집계
tag2=count.most_common(20)  # 상위20개만추출
print(tag2,type(tag2))
taglist=list(tag2)  # tag2 데이터로 태그 list 생성
print(taglist)
f.close()
```
